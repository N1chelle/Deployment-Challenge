```python
games_df = df.loc[:,['name','required_age', 'is_free','review_score']]
games_df = games_df.head(10)
games_df
```

```python
# Importing sql library

from sqlalchemy import create_engine
```

```python
# Create a reference for sql library

connection_string = 'sqlite:///game.db' 
engine = create_engine(connection_string, echo=False)
```

```python
# Attach the data frame to the sql
games_df.to_sql('Gaming_Data',
                con = engine, if_exists="replace")
```

```python
pd.read_sql('SELECT* FROM Gaming_Data', con = engine)
```
